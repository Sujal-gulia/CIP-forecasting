```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv('50 Macroeconomic Indicators.csv')
df.head()
```

```python
df.replace('-', pd.NA, inplace=True)
```

```python
df.isnull().sum()
```

```python
print(df.columns.to_list)
```

```python
df.rename(columns ={
    'Commercial Paper (Amount Outstanding) \r\n(? Crore)' : 'Commercial Paper (Amount Outstanding) (in Crore)',
    'Net Foreign Direct Investment \r\n(US $ Million)' : 'Net Foreign Direct Investment (US $ Million)',
    'Direct Investment to India \r\n(US $ Million)' : 'Direct Investment to India (US $ Million)',
    'Foreign Direct Investment By India\r\n (US $ Million)' : 'Foreign Direct Investment By India (US $ Million)',
    'Net Portfolio Investment \r\n(US $ Million)' : 'Net Portfolio Investment (US $ Million)',
    'Total Investment Inflows \r\n(US $ Million)' : 'Total Investment Inflows (US $ Million)',
    'FCNR(B)-Foreign Currency Non-Resident (Banks) \r\n(US $ Million)' : 'FCNR(B)-Foreign Currency Non-Resident (Banks) (US $ Million)',
    'External Commercial Borrowings � Registrations\r\n(US $ Million)' : 'External Commercial Borrowings Registrations (US $ Million)',
    'Foreign Trade Exports Total \r\n(US $ Million)' : 'Foreign Trade Exports Total (US $ Million)',
    'Foreign Trade Imports Total \r\n(US $ Million)' : 'Foreign Trade Imports Total (US $ Million)',
    'Foreign Trade Balance Total \r\n(US $ Million)' : 'Foreign Trade Balance Total (US $ Million)',
    'Market Borrowing SG Gross Amount Raised \r\n(? Crore)' : 'Market Borrowing SG Gross Amount Raised (in Crore)',
    'Exchange Rate of Indian Rupee vis-�-vis US Dollar (Month End)' : 'Exchange Rate of Indian Rupee to US Dollar (Month End)'
}, inplace = True)

df.head()
```

```python
#dropping total retail payments, and total digital paymnets columns as they have most of the values as null values
df.drop(['Total Retail Payments \r\n(? Crore)', 'Total Digital Payments \r\n(? Crore)'], axis = 1, inplace=True)
df.head()
```

```python
#dropping first two rows as they have null values for most of the columns
df.drop(index=[0, 1], inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
pd.to_datetime(df['Period'], format='%b-%Y')
df.set_index('Period', inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# converting object columns to numeric
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = pd.to_numeric(df[col].str.replace(',', '').str.strip(), errors='coerce')
```

```python
df['Consumer Price Index  (2012=100)'].plot()
df['Index of Industrial Production'].plot()
```

```python
df['Market Borrowing SG Gross Amount Raised (in Crore)'].plot()
```

```python
columns = ['Consumer Price Index  (2012=100)', 'Index of Industrial Production', 'Market Borrowing SG Gross Amount Raised (in Crore)']

forward_mean = df[columns].fillna(method='ffill').rolling(window=3, min_periods=1).mean()
backward_mean = df[columns].fillna(method='bfill').rolling(window=3, min_periods=1).mean()

df[columns] = df[columns].fillna((forward_mean + backward_mean) / 2)
```

```python
df['Consumer Price Index  (2012=100)'].plot()
df['Index of Industrial Production'].plot()
```

```python
df['Market Borrowing SG Gross Amount Raised (in Crore)'].plot()
```

```python
corr_matrix = df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
```
